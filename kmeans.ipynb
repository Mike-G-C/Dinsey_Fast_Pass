```python
import pandas as pd
df = pd.read_csv('Paris historical data with FastPass.csv')
```

```python
df
```

```python
x =df[['Income','Merchandise Spend']]
l = []
c=[]
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn import metrics



k_means = KMeans(n_clusters=2, random_state=10)

k_means.fit(x)

y_predict_2 = k_means.predict(x)
print(k_means.inertia_)
l.append(k_means.inertia_)
c.append(k_means.cluster_centers_)
```

```python
k_means = KMeans(n_clusters=3)

k_means.fit(x)

y_predict_3 = k_means.predict(x)
print(k_means.inertia_)
l.append(k_means.inertia_)
c.append(k_means.cluster_centers_)
```

```python
k_means = KMeans(n_clusters=4)

k_means.fit(x)

y_predict_4 = k_means.predict(x)
print(k_means.inertia_)
l.append(k_means.inertia_)
c.append(k_means.cluster_centers_)
```

```python
k_means = KMeans(n_clusters=5)

k_means.fit(x)

y_predict_5 = k_means.predict(x)
print(k_means.inertia_)
l.append(k_means.inertia_)
c.append(k_means.cluster_centers_)
```

```python
k_means = KMeans(n_clusters=6)

k_means.fit(x)

y_predict_6 = k_means.predict(x)
print(k_means.inertia_)
l.append(k_means.inertia_)
c.append(k_means.cluster_centers_)
```

```python
import matplotlib.pyplot as plt
plt.plot([2,3,4,5,6],l,'o-')
plt.show()
```

```python
df['cluster2']=y_predict_2
df['cluster3']=y_predict_3
df['cluster4']=y_predict_4
df['cluster5']=y_predict_5
df['cluster6']=y_predict_6

df
```
